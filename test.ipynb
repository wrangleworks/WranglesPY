```python
import wrangles
import pandas as pd
```

```python
dataframe=pd.DataFrame({
    'col1': ['a'],
    'col2': ['b'],
    'col3': ['c']
})
display(dataframe)
```

```python
# Rename with simple mapping
recipe = '''
wrangles:
- rename:
    col1: new_col1
    col2: new_col2
'''
display(dataframe)
display(wrangles.recipe.run(recipe, dataframe=dataframe))
```

```python
# Rename with input/output lists, (one extra optional column that doesn't exist)
dataframe=pd.DataFrame({
    'col1': ['a'],
    'col2': ['b']
})
recipe = '''
wrangles:
- rename:
    col1?: new_col1
    missing_col?: new_missing
    col2: new_col2
'''
display(dataframe)
display(wrangles.recipe.run(recipe, dataframe=dataframe))
```

```python
# Rename with optional columns only that don't exist
dataframe=pd.DataFrame({
    'col1': ['a'],
    'col2': ['b']
})
recipe = '''
wrangles:
- rename:
    missing1?: new1
    missing2?: new2
'''
display(dataframe)
display(wrangles.recipe.run(recipe, dataframe=dataframe))
```

```python
# Rename with input/output format and optional columns (all exist)
dataframe=pd.DataFrame({
    'col1': ['a'],
    'col2': ['b'],
    'col3': ['c']
})
recipe = '''
wrangles:
- rename:
    input:
        - col1?
        - col2?
        - col3
    output:
        - new_col1
        - new_col2
        - new_col3
'''
display(dataframe)
display(wrangles.recipe.run(recipe, dataframe=dataframe))
```

```python

# Rename with input/output format and optional columns (all exist)
dataframe=pd.DataFrame({
    'col1': ['a'],
    'col2': ['b']
})
recipe = '''
    wrangles:
    - rename:
        input:
            - first?
            - second?
        output:
            - Two
            - Columns
'''
display(dataframe)
display(wrangles.recipe.run(recipe, dataframe=dataframe))
```

```python
# 7. Rename with mix of optional and required columns

dataframe=pd.DataFrame({
    'required_col': ['a'],
    'optional1': ['b'],
    'another_required': ['c']
})
recipe = '''
    wrangles:
    - rename:
        required_col: new_required
        optional1?: new_opt1
        optional2?: new_opt2
        another_required: new_another
'''
display(dataframe)
display(wrangles.recipe.run(recipe, dataframe=dataframe))
```

```python
# 8. Rename with required column missing (should raise error)

dataframe=pd.DataFrame({
    'other_col': ['a']
})
recipe = '''
    wrangles:
    - rename:
        required_col: new_required
        optional?: new_opt
'''
display(dataframe)
display(wrangles.recipe.run(recipe, dataframe=dataframe))
```

```python
# 9. Rename with empty dataframe (optional columns)

dataframe=pd.DataFrame()
recipe = '''
    wrangles:
    - rename:
        col1?: new_col1
        col2?: new_col2
'''
display(dataframe)
display(wrangles.recipe.run(recipe, dataframe=dataframe))
```

```python
# Rename with where clause and optional column (should raise NotImplementedError)

dataframe=pd.DataFrame({
    'col1': [1, 6],
    'col2': [2, 7]
})
recipe = '''
wrangles:
- rename:
    col1?: new_col1
    col2: new_col2
    where: col1 > 5
'''
display(dataframe)
display(wrangles.recipe.run(recipe, dataframe=dataframe))
```
